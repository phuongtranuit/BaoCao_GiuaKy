<h3>2. Thực hiện Code</h3>

<b>Chèn thư viện mtcnn, cv2 vào chương trình</b>



```python
#import các thư viện và packages
from mtcnn.mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
import os
```

<b>Đọc hình gốc imgdemo.png từ hệ thống</b>

```python
img = cv2.imread("imgdemo.png") #Đọc ảnh img1.jpg lưu vào biến img
temp = img.copy() #Tạo ra 1 bản sao của biến img lưu vào temp
```

<b>Duyệt, tách từng face và đưa vào thư mục tương ứng</b>
-	Duyệt từng vị trí khuôn mặt.<br />
-	Lưu ảnh vào mảng temp để xử lý chuyển chế độ bgr sang rgb để vẽ<br />
-	Phát hiện sẽ vẽ hình chữ nhật màu xanh lên khuôn mặt đã được phát hiện.<br />
-	Tạo đường dẫn thư mục để chứa các khuôn mặt được cắt nhỏ từ ảnh gốc


```python
#Tạo biến d như là index để lưu tên ảnh (img1, img2,...)
d = 1 
#print(detector.detect_faces(img))
#cho biến i chạy qua các kết quả output
for i in out:
    for[x, y, w, h] in [i['box']]:
        print(x,y,w,h)#in ra tọa độ chiều rộng chiều cao của các output
        cv2.rectangle(img, (x,y),(x+w,y+h), (255,0,0),3)#vẽ tứ giác viền màu xanh xung quanh
        sub_face = temp[y:y+h,x:x+w]#Lưu sub_face vào mảng temp
        plt.imshow(cv2.cvtColor(sub_face, cv2.COLOR_BGR2RGB))#Chuyển màu sub_face từ bgr sang rgb và Vẽ ảnh sub_face
        plt.show()#Hiển thị ảnh
        directory = "face"+str(d)
        if not os.path.exists(directory):#Nếu không tồn tại đường dẫn
            os.makedirs(directory)#Tạo đường dẫn
            cv2.imwrite("./face"+str(d)+"/"+"face"+str(d)+".jpg",sub_face)#Write hình xuống đường dẫn (/face1/face1.jpg)
        d+=1
```
